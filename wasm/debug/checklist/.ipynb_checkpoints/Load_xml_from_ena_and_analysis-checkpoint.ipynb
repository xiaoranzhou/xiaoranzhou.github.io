# Read XML and get mandatory 

## load xml from file system


```python
import os
import xml.etree.ElementTree as ET

names = os.listdir()
```

```python
file_list = list(filter(lambda name: name[0:3] == "ERC", names ))
```

```python
for index, name in enumerate(file_list):
    tree = ET.parse(name)
    root = tree.getroot()

    label= root.findall(".//DESCRIPTOR/LABEL")
    
    print(name , "  ",label[0].text)
```
